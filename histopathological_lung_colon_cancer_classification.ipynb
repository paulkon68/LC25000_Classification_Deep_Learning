Η συνάρτηση `create_sets()` χρησιμοποιείται για την δημιουργία των συνόλων δεδομένων εκπαίδευσης, επαλήθευσης και δοκιμής. Πιο συγκεκριμένα, αντιγράφει αρχεία από έναν αρχικό φάκελο `source_dir`, στον φάκελο προορισμού που παρέχεται με το όρισμα `destination_dir`. Τα πρώτα τρία ορίσματα, υποδηλώνουν την μορφή του ονόματος του αρχείου (`file_format`), καθώς και το πλήθος των αρχείων που θα αντιγραφούν στον φάκελο: `rngs`, `rnge`. 

Για παράδειγμα, η μορφή των αρχείων εικόνων του αδενοκαρκινώματος του πνεύμονα είναι `lungaca{}.jpeg`, όπου αντί για τις αγκύλες παρατηρείται ένα αριθμός από το 1 μέχρι και το 5000 (αρχικός φάκελος δεδομένων). Αν θέλουμε να αντιγράψουμε τις πρώτες 2500 εικόνες, ορίζουμε τους φακέλους πηγής (source) και προορισμού (destination), και ως file_format το `lungaca{}.jpeg`. Τέλος, ως `rngs` θέτουμε το 1 και ως `rnge` το 2501 (διότι η συνάρτηση `range()`, επιστρέφει ένα αντικείμενο που είναι `iterable` από το 1 εως και το 2500, το 2501 δεν περιλαμβάνεται).

```python
import shutil

 
def create_sets(file_format, rngs, rnge, source_dir, destination_dir):

  # Δημιούργησε μια λίστα με τα ονόματα των αρχείων που θα αντιγραφούν.
  fnames = [f'{file_format}'.format(i) for i in range(rngs, rnge)]

  # Για κάθε αρχείο (εικόνα) στην λίστα, 
  for fname in fnames:
      src = os.path.join(source_dir, fname)  # το μονοπάτι της πηγής των δεδομένων 
      dst = os.path.join(destination_dir, fname)  # το μονοπάτι στον φάκελο όπου και θα αντιγραφεί το αρχείο
      
      if not os.path.exists(dst):
          # αντέγραψε το αρχείο από τον φάκελο src στον φάκελο dst.
          shutil.copyfile(src, dst)
```

Ο παρακάτω κώδικας, χρησιμοποιεί την συνάρτηση `create_sets()` που περιγράφηκε προηγουμένως, για να δημιουργήσει το σετ δεδομένων εκπαίδευσης (training set). Όπως αναφέρθηκε και προηγουμένως, θα αντιγραφούν 2500 εικόνες για κάθε κλάση στον αντίστοιχο υποφάκελο που βρίσκεται μέσα στον φάκελο `training/` .

```python
img_width = 256  # όρισε το πλάτος των εικόνων
img_height = 256  # όρισε το ύψος των εικόνων
batch_size = 128  # όρισε το μέγεθος της ομάδας παρατηρήσεων 
```

Στα δεδομένα εκπαίδευσης, εφαρμόστηκε η τεχνική της επαύξησης των δεδομένων `Data Augmentation`. Η τεχνική αυτή, δημιουργεί παραλλαγές τις αρχικής εικόνας παρέχοντας μια νέα προοπτική αποτύπωσης του αντικειμένου στην πραγματική ζωή. Αναλυτικότερα, οι εικόνες που παράγονται από την εφαρμογή της τεχνικής είναι αποτέλεσμα μετασχηματισμών όπως: μετατόπιση, περιστροφή, κατοπτρισμος κ.α.

Η τεχνική της επαύξησης των δεδομένων πραγματοποιήθηκε με τη βοήθεια της κλάσης [`ImageDataGenerator`](https://keras.io/api/preprocessing/image/) του Keras, η οποία αποτελεί έναν γρήγορο και εύκολο τρόπο για την αύξηση των δεδομένων. Το βασικό πλεονέκτημα αυτής της κλάσης είναι ότι έχει σχεδιαστεί για να παρέχει αύξηση δεδομένων σε πραγματικό χρόνο. Αυτό σημαίνει ότι δημιουργεί επαυξημένες εικόνες εν κινήσει ενώ το μοντέλο βρίσκεται στο στάδιο της εκπαίδευσης. Επιπλέον, διασφαλίζει ότι το μοντέλο λαμβάνει νέες παραλλαγές των εικόνων σε κάθε εποχή, επιστρέφοντας ταυτόχρονα μόνο τις μετασχηματισμένες εικόνες χωρίς αυτές να προστίθενται στο αρχικό σώμα των εικόνων.

Ένα άλλο πλεονέκτημα του ImageDataGenerator είναι ότι έχει λιγότερες απαιτήσεις σε μνήμη. Αυτό συμβαίνει γιατί χωρίς να χρησιμοποιήσουμε αυτήν την κλάση, φορτώνουμε όλες τις εικόνες ταυτόχρονα, ενώ όταν την χρησιμοποιούμε, φορτώνουμε τις εικόνες σε ομάδες δεδομένων (batches).

Στην συγκεκριμένη περίπτωση, οι εικόνες υφίστανται τους παρακάτω μετασχηματισμους: 

- οι εικόνες θα πρέπει να κανονικοποιηθούν, και συγκεκριμένα να επανακλιμακοποιηθούν στο μοναδιαίο εύρος, δηλαδή στο διάστημα [0,1], καθώς τα νευρωνικά δίκτυα είναι ευαίσθητα στην κλίμακα των εισόδων. (rescale=1/255.0)

- περιστροφή εως 30 μοίρες (rotation_range=30)

- μεγένθυνση εως 40% (zoom_range=0.4)

- οριζόντιος κατοπτρισμός (horizontal_flip=True)

Η μέθοδος `flow_from_directory()`, λαμβάνει ένα μονοπάτι προς τα δεδομένα και παράγει ομάδες μετασχηματισμένων εικόνων (augmented batches of images). 

Το αποτέλεσμα είναι μια ομάδα παρατηρήσεων ίση με το `batch_size` που ορίστηκε προηγουμένως και της σχετικές ετικέτες τους. Οι εικόνες θα έχουν το σχήμα (batch_size, img_width, img_height, channels), ενώ οι ετικέτες τους θα έχουν σχήμα: (batch_size, 1).

```python
# Ορίζουμε τους μετασχηματισμούς που θα εφαρμοστούν κατά την επαύξηση των δεδομένων εκπαίδευσης
train_datagen = ImageDataGenerator(rescale = 1/255.0, # Η είσοδος του νευρωνικού είναι ευαίσθητη 
                                   rotation_range=30,  # περιστροφή εως 30 μοίρες
                                   zoom_range=0.4,  # μεγένθυνση εως 40%
                                   horizontal_flip=True  # οριζόντιος κατοπτρισμός
                                   )

# Δημιούργησε τον generator των επαυξημένων δεδομένων 
train_generator = train_datagen.flow_from_directory(training_set_dir,  # όρισε το μονοπάτι προς τα δεδομένα
                                                    batch_size=batch_size,  # όρισε το μέγεθος των παρατηρήσεων
                                                    class_mode='categorical',  # one-hot encoding 
                                                    target_size=(img_height, img_width)  # όρισε τις διαστάσεις των εικόνων
                                                    )
```

Με όμοιο τρόπο χρησιμοποιούμε την κλάση `ImageDataGenerator` και την μέθοδο `flow_from_directory` για τα δεδομένα επαλήθευσης. Ωστόσο, ο μόνος μετασχηματισμός που εφαρμόζουμε στις εικόνες είναι αυτός της επανακλιμάκωσης των εικόνων στο μοναδιαίο εύρος.

```python
validation_datagen = ImageDataGenerator(rescale = 1/255.0)

validation_generator = validation_datagen.flow_from_directory(validation_set_dir,
                                                              batch_size=batch_size,
                                                              class_mode='categorical',
                                                              target_size=(img_height, img_width)
                                                             )
```

Για την επίλυση του προβλήματος ταξινόμησης των εικόνων του συνόλου δεδομένων LC25000, θα σχεδιαστεί ένα Συνελικτικό Νευρωνικό Δίκτυο (Convolutional Neural Network - CNN). Τα επίπεδα της αρχιτεκτονικής θα συνδέονται με ακολουθιακό τρόπο, και συνεπώς το μοντέλο αυτό μπορεί να κατασκευαστεί δημιουργώντας ένα Ακολουθιακό Μοντέλο (Sequential Model).

Χρησιμοποιήθηκαν τα παρακάτω είδη επιπέδων:

- Συνελικτικό επίπεδο 2 διαστάσεων (Conv2D)
- Επίπεδο Μέγιστης Ομαδοποίησης (MaxPooling2D)
- Επίπεδο Flatten
- Πλήρως Συνδεδεμένο Επίπεδο (Dense/Fully Connected)

Σε όλα τα επίπεδα ως συνάρτηση ενεργοποίησης ορίστηκε η Rectified Linear Unit (ReLU), με εξαίρεση το τελευταίο.
Το τελευταίο επίπεδο είναι το επίπεδο εξόδου (output layer), το οποίο διαθέτει 5 νευρώνες (αφού έχουμε 5 κλάσεις στο σύνολο δεδομένων), ενώ ως συνάρτηση ενεργοποίησης ορίστηκε η softmax, η οποία επιτρέπει την αντιστοίχιση των εξόδων του προηγούμενου επιπέδου σε πιθανότητες συμμετοχής σε κλάσεις (class membership probabilities), για το πρόβλημα ταξινόμησης των 5 κλάσεων. 

Με την χρήση της μεθόδου `summary()`, μπορούμε να δούμε αν το μοντέλο μας έχει δημιουργηθεί σωστά.

```python
model = Sequential([
    
    # 1ο επίπεδο
    Conv2D(16,  # πλήθος συνελικτικών φίλτρων
           (3, 3),  # μέγεθος διδιάστατου παραθύρου συνέλιξης (πυρήνας-kernel)
           activation='relu',  # ορισμός συνάρτησης ενεργοποίησης
           input_shape=(img_height, img_width, 3)),  # ορισμός σχήματος εισόδου

    # 2ο επίπεδο
    MaxPooling2D(2, 2),  #  Εφάρμοσε Max-Pooling με υποδιπλασιασμό κάθε διάστασης
    
    # 3ο επίπεδο
    Conv2D(32, 
           (3, 3), 
           activation='relu'), 

    # 4ο επίπεδο
    MaxPooling2D(2, 2),  #  Εφάρμοσε Max-Pooling με υποδιπλασιασμό κάθε διάστασης

    # 5ο επίπεδο
    Conv2D(64, 
           (3, 3), 
           activation='relu'),

    # 6ο επίπεδο
    Conv2D(64, 
           (3, 3), 
           activation='relu'),

    # 7ο επίπεδο
    MaxPooling2D(2, 2),  #  Εφάρμοσε Max-Pooling με υποδιπλασιασμό κάθε διάστασης

    # 8ο επίπεδο
    Conv2D(128, 
           (3, 3), 
           activation='relu'),

    # 9ο επίπεδο
    Conv2D(128, 
           (3, 3), 
           activation='relu'),

    # 10ο επίπεδο
    MaxPooling2D(2, 2),  #  Εφάρμοσε Max-Pooling με υποδιπλασιασμό κάθε διάστασης

    # 11ο επίπεδο
    Conv2D(256, 
           (3, 3), 
           activation='relu'),

    # 12ο επίπεδο
    Conv2D(256, 
           (3, 3), 
           activation='relu'),

    # 13ο επίπεδο
    Conv2D(256, 
           (3, 3), 
           activation='relu'),

    # 14ο επίπεδο
    MaxPooling2D(2, 2),  #  Εφάρμοσε Max-Pooling με υποδιπλασιασμό κάθε διάστασης

    # 15ο επίπεδο
    Flatten(),  # Κάνε την έξοδο του προηγούμενου επιπέδου διάνυσμα

    # 16ο επίπεδο
    Dense(512, 
          activation='relu'),

    # 17ο επίπεδο
    Dense(512, 
          activation='relu'),

    # 18ο επίπεδο - επίπεδο εξόδου
    Dense(5,  # 5 κλάσεις, άρα 5 νευρώνες στην έξοδο
          activation='softmax')  # Συνάρτηση ενεργοποίησης για Ταξινόμηση πολλών κλάσεων 
])

model.summary()
```

Στην συνέχεια, προστίθενται στο μοντέλο οι λεπτομέρειες αναφορικά με τον Αλγόριθμο Προσαρμογής (Fitting Algorithm). Ειδικότερα, θα πραγματοποιηθεί ελαχιστοποίηση της Συνάρτησης Σφάλματος Cross-Entropy (Cross-Entropy Loss Function), ενώ ο Βελτιστοποιητής (Optimizer), που θα χρησιμοποιηθεί είναι ο Adam. Παράλληλα κατά την διάρκεια της εκπαίδευσης και της επαλήθευσης θα υπολογίζεται και η μετρική της Ακρίβειας (Accuracy).

```python
model.compile(optimizer='Adam', 
              loss='categorical_crossentropy', 
              metrics =['accuracy'])
```

Αφού σχεδιαστεί η αρχιτεκτονική του μοντέλου, είναι δυνατή η έναρξη της εκπαίδευσης του Συνελικτικού Νευρωνικού Δίκτου που κατασκευάστηκε.

Η διαδικασία της εκπαίδευσης του δικτύου πραγματοποιείται με τη μέθοδο `fit` του μοντέλου που κατασκευάστηκε. Ως πρώτο όρισμα της μεθόδου, δινεται ο generator των batches των εικόνων εκπαίδευσης που δημιουργήθηκε με την χρήση της κλάσης `ImageDataGenerator`.
```python
# Εισαγωγή της βιβλιοθήκης pandas με το ψευδώνυμο pd
import pandas as pd


# Δημιουργία της λίστα που θα περιέχει τις ετικέτες των εικόνων, 
# όπως οι τελευταίες συναντώνται μέσα στον φάκελο "test_images/"
label_list = []

# Δημιουργία λεξικού με κλειδιά τις κλάσεις και τιμές το είδος των εικόνων.
classes_classification = {0: "colonca", 
                          1: "colonn", 
                          2: "lungaca", 
                          3: "lungn", 
                          4: "lungscc"}

# Δημιούργησε μια λίστα με τα ονόματα των αρχείων (εικόνες) που βρίσκονται μέσα 
# στον φάκελο "test_images/"
image_id = [i for i in os.listdir("dataset/test_images")]

# Αποθήκευσε τις σωστές ετικέτες στην λίστα 
label_list = [k for i in image_id for k, v in classes_classification.items() if v in i ]

# Δημιούργησε ένα DataFrame, με στήλες τα ονόματα των αρχείων και τις 
# αντίστοιχες ετικέτες τους
labels = {
  'Image_id': image_id,
  'labels': label_list
}

df = pd.DataFrame(labels)

# Αποθήκευσε το DataFrame στο αρχείο με όνομα "labels.csv"
df.to_csv("dataset/labels.csv")
```

Η συνάρτηση `preprocess_image`, λαμβάνει ως όρισμα το μονοπάτι προς μια εικόνα. Αφού φορτώσει τις τιμές της εικόνας σε μια μεταβλητή με την βοήθεια της συνάρτησης `img_to_array`, δημιουργεί εναν πίνακα διαστάσεων 256x256x3. Έπειτα, μετατρέπεται αυτός ο πίνακας σε τετραδιάστατο με διαστάσεις 1x256x256x3. Η πρώτη διάσταση του τετραδιάστατου πινακα δείχνει στην μέθοδο `predict` του μοντέλου ότι οι εικόνες θα προβλέπονται σε ομάδες (batches) της μιας εικόνας.

```python
# Εισαγωγή των απαραίτητων βιβλιοθηκών που θα χρησιμοποιηθούν από την συνάρτηση "preprocess_image"
import numpy as np
from keras.preprocessing.image import load_img, img_to_array

# Η παρακάτω συνάρτηση θα διαβάζει μια εικόνα την φορά από το μονοπάτι 
# που θα δίνεται ως όρισμα και θα επανακλιμακώνει τις τιμές των εικόνων στο 
# διάστημα [0, 1]
def preprocess_image(path):
    img = load_img(path, target_size = (img_height, img_width))
    a = img_to_array(img)  # δημιούργησε τον πίνακα με τις τιμές της εικόνας
    a = np.expand_dims(a, axis = 0)  # μετέτρεψε τον τρισδιάστατο πινακα σε τετραδιάστατο
    a /= 255.  # επανακλιμάκωσε τις τιμές στο διάστημα [0, 1]
    return a  # επέστρεψε την επεξεργασμένη εικόνα
```

Με την εκτέλεση του παρακάτω κελιού, θα δημιουργηθεί ένας πίνακας που θα περιέχει τις επεξεργασμένες εικόνες δοκιμής, δηλαδή ένας πίνακας πινακων, οι οποίοι θα βρίσκονται ο ένας κάτω από τον άλλο. Αυτός ο πίνακας θα δημιουργηθεί με τη βοήθεια της βιβλιοθήκης numpy και ειδικότερα, με την συνάρτηση `vstack`, η οποία συνενώνει πίνακες με τον τρόπο που περιγράφηκε προηγουμένως.

```python
# Όρισε ως μεταβλητές τον φάκελο των εικόνων δοκιμής και το αρχείο των ετικετών τους
test_images_dir = 'dataset/test_images/'
test_df = pd.read_csv('dataset/labels.csv')

# Δημιούργησε μια λίστα με τα ονόματα των εικόνων που βρίσκονται στην στήλη
# 'Image_id' του αρχείου csv.
test_dfToList = test_df['Image_id'].tolist()
test_ids = [str(item) for item in test_dfToList]

# Δημιούργησε μια λίστα με τα μονοπάτια όλων των εικόνων που πρόκειτα να προβλεφθούν
test_images = [test_images_dir+item for item in test_ids]

# Αποθήκευσε τις επανακλιμακωμένες εικόνες σε μια λίστα 
prep_imgs = [preprocess_image(fn) for fn in test_images]

# Αποθήκευσε τις εικόνες σαν πίνακες, τον έναν κάτω από τον άλλο
test_preprocessed_images = np.vstack(prep_imgs)
```

Στο σημείο αυτό, μέσω της μεθόδου `predict`, θα προβλεφθούν σε ποιές κλάσεις ανήκουν οι 6250 εικόνες του συνόλου δοκιμής, σε ομάδες (batches) της μιας εικόνας τη φορά.  

```python
# Πρόβλεψε τις πιθανότητες  συμμετοχής των εικόνων δοκιμής στις κλάσεις
array = model.predict(test_preprocessed_images, batch_size=1, verbose=1)

# Ανέθεσε την κάθε εικόνα στην κλάση με την μεγαλύτερη πιθανότητα
answer = np.argmax(array, axis=1)
```
